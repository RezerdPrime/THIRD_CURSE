```python
import numpy as np
from matplotlib import pyplot as plt

def f(x1, x2, x3):
    return x1**2 + 2*x2**2 + 3*x3**2 + 2*x1*x3 + 2*x2*x3 + 2*(x1+x2+x3)

def gradf(x1, x2, x3):
    return np.array([2*x1 + 2*x3 + 2, 4*x2 + 2*x3 + 2, 6*x3 + 2*x1 + 2*x2 + 2])

def VectorNorm2(V):
    return sum([v**2 for v in V])**.5

def GradientDescent(x0, h0, eps):
    h = h0
    xcur = x0
    gf = gradf(*xcur)
    xnext = xcur - h*gf

    count = 0

    while VectorNorm2(gf) > eps:
        if f(*xnext) >= f(*xcur): h /= 2
        
        xcur = xnext
        gf = gradf(*xcur)
        xnext = xcur - h * gf

        count += 1

    return xnext, count
```
