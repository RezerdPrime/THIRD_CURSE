```python
import numpy as np

def Norm2(v):
    return sum([v_[0]*v_[0] for v_ in v])**.5
```

# Метод Якоби

```python
def JacobiMethod(A_, b_, x0, eps):
    n = len(A_)
    A = np.zeros((n,n))
    b = np.zeros((n,1))

    kaunt = 0

    for i in range(n):
        b[i, 0] = b_[i, 0] / A_[i, i]

        for j in range(n):
            if i != j: A[i, j] = -A_[i, j] / A_[i, i]

    xbuf = np.zeros((n,1))
    xnext = x0

    while Norm2(xnext - xbuf) > eps:
        xbuf = xnext
        xnext = A @ xnext + b

        kaunt += 1
        #print(xnext)

    return xnext, kaunt


A = np.array([
    [4, 2, 1],
    [2, 4, 1],
    [1, 2, 4]
])

b = np.array([
    [1],
    [1],
    [1]
])

x = np.array([
    [-1],
    [1],
    [0]
])


x_, k = JacobiMethod(A, b, x, 0.001)

print(Norm2(x_ - np.array([[1/7], [1/7], [1/7]])), k)
print()


A = np.array([
    [6.22, 1.42, -1.72, 1.91],
    [1.42, 5.33, 1.11, -1.82],
    [-1.72, 1.11, 5.24, 1.42],
    [1.91, -1.82, 1.42, 6.55]
])

b = np.array([
    [7.53],
    [6.06],
    [8.05],
    [8.06]
])

x = np.array([
    [-1],
    [1],
    [1],
    [0]
])


x_, k = JacobiMethod(A, b, x, 0.0001)

print(x_, k)
print(Norm2(A @ x_ - b))
```

# Метод верхней релаксации

```python
def URM(A_, b_, x0, omega, eps):
    n = len(A_)
    
    D = np.zeros((n,n))
    U = np.zeros((n,n))
    L = np.zeros((n,n))

    kaunt = 0

    for i in range(n):

        for j in range(n):
            if i > j: 
                U[i, j] = A_[i, j]
            elif i < j:
                L[i, j] = A_[i, j]
            else:
                D[i, j] = A_[i, j]

    Lowerinv = np.linalg.inv(D + omega * L)
    bnew = omega * b_
    Upper = omega * U + (omega - 1) * D

    xbuf = np.zeros((n,1))
    xnext = x0

    while Norm2(xnext - xbuf) > eps:
        xbuf = xnext
        xnext = Lowerinv @ (bnew - Upper @ xnext)
        kaunt += 1

    return xnext, kaunt


A = np.array([
    [4, 2, 1],
    [2, 4, 1],
    [1, 2, 4]
])

b = np.array([
    [1],
    [1],
    [1]
])

x = np.array([
    [-1],
    [1],
    [0]
])

x_, k = URM(A, b, x, 1.1, 0.001)

print(Norm2(x_ - np.array([[1/7], [1/7], [1/7]])), k)
print()

A = np.array([
    [6.22, 1.42, -1.72, 1.91],
    [1.42, 5.33, 1.11, -1.82],
    [-1.72, 1.11, 5.24, 1.42],
    [1.91, -1.82, 1.42, 6.55]
])

b = np.array([
    [7.53],
    [6.06],
    [8.05],
    [8.06]
])

x = np.array([
    [-1],
    [1],
    [1],
    [0]
])

x_, k = URM(A, b, x, 1.1, 0.0001)

print(x_, k)
print(Norm2(A @ x_ - b))
print()

for i in range(100):
    x_, k = URM(A, b, x, 1 + i/100, 0.0001)
    print(f"ω: {1 + i/100}\t\t k: {k}")
```
