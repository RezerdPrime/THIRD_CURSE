```python
import numpy as np
import pandas as pd
pd.set_option('display.float_format', '{:.3f}'.format)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)

N = 25
normsA = []
normsB = []
conds = []


def MatrixNorm(M): 
    return max([sum(M[i]) for i in range(N)])

def VectorNorm2(V):
    return (sum([v**2 for v in V]))**.5


for i in range(5):

    A = np.random.rand(N, N)
    B = np.linalg.inv(A)
    normA = MatrixNorm(A)
    normB = MatrixNorm(B)

    normsA.append(normA)
    normsB.append(normB)
    conds.append(normA*normB)

    print("\n\n")
    print(f'matrix #{i+1}')
    print(pd.DataFrame(A))
```

```python
print(f"№\t\t||A||\t\t\t\t||A^-1||\t\t\tcond(A)")
print("="*100)

for i in range(5):
    print(f"{i+1}\t\t{normsA[i]}\t\t{normsB[i]}\t\t{conds[i]}")
```

```python
a = 0; b = 1
xs = np.linspace(a, b, N)

exactsol = np.array([1 for _ in range(N)]).T

W = np.array([[xs[i]**j for j in range(N)] for i in range(N)])
y = W @ exactsol

#x = np.linalg.lstsq(W, y, rcond=None)[0]
x = np.linalg.solve(W, y)

print(MatrixNorm(W)*MatrixNorm(np.linalg.inv(W)), VectorNorm2(x - exactsol))

x
```
