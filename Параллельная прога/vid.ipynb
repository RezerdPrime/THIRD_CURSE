```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import os

#    
def read_data_from_file(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    
    points = []
    for line in lines:
        if line.strip():  #   
            point = list(map(float, line.split()))
            points.append(point)
    
    return np.array(points)

#     
filename = 'pos.txt'
points = read_data_from_file(filename)

#     
output_dir = 'output_images'
os.makedirs(output_dir, exist_ok=True)  #  ,    

#    
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#   
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
ax.set_title('3D Visualization of Points')

#       
for i in range(0, len(points), 4):
    if i + 4 <= len(points):
        current_points = points[i:i+4]
        ax.cla()  #  
        ax.set_xlim([-1, 1])
        ax.set_ylim([-1, 1])
        ax.set_zlim([-1, 1])
        ax.set_title('3D Visualization of Points')

        #    3D 
        ax.scatter(current_points[:, 0], current_points[:, 1], current_points[:, 2], s=100)  # s -  

        #  
        plt.savefig(os.path.join(output_dir, f'frame_{i//4:03d}.png'))  #   
        plt.pause(0.5)  #   

plt.show()
```
