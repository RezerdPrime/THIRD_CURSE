```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import os

# Считывание данных из файла
def read_data_from_file(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    
    points = []
    for line in lines:
        if line.strip():  # Пропускаем пустые строки
            point = list(map(float, line.split()))
            points.append(point)
    
    return np.array(points)

# Укажите путь к вашему файлу
filename = 'pos.txt'
points = read_data_from_file(filename)

# Создание директории для хранения изображений
output_dir = 'output_images'
os.makedirs(output_dir, exist_ok=True)  # Создание директории, если она не существует

# Визуализация и сохранение изображений
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Устанавливаем параметры осей
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
ax.set_title('3D Visualization of Points')

# Функция для анимации покадрово и сохранения изображений
for i in range(0, len(points), 4):
    if i + 4 <= len(points):
        current_points = points[i:i+4]
        ax.cla()  # Очищаем оси
        ax.set_xlim([-1, 1])
        ax.set_ylim([-1, 1])
        ax.set_zlim([-1, 1])
        ax.set_title('3D Visualization of Points')

        # Добавляем точки в 3D пространство
        ax.scatter(current_points[:, 0], current_points[:, 1], current_points[:, 2], s=100)  # s - размер точек

        # Сохранение изображения
        plt.savefig(os.path.join(output_dir, f'frame_{i//4:03d}.png'))  # Формат имени файла
        plt.pause(0.5)  # Пауза между кадрами

plt.show()
```
